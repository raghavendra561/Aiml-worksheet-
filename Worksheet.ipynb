```python
students = []

# Add 3 students
for i in range(3):
    roll = int(input("Enter roll: "))
    name = input("Enter name: ")
    m1 = int(input("Mark 1: "))
    m2 = int(input("Mark 2: "))
    m3 = int(input("Mark 3: "))

    record = {
        "roll": roll,
        "name": name,
        "marks": [m1, m2, m3]
    }

    students.append(record)

# Print each student name and average mark
for s in students:
    marks = s["marks"]
    avg = sum(marks) / len(marks)

    if avg >= 50:
        result = "Pass"
    else:
        result = "Fail"

    print(s["roll"], s["name"], "Average:", avg, result)
```
