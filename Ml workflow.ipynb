```python
import pandas as pd
# Sample farm dataset
data = {
 "Rainfall_mm": [200, 190, 220, 280, 190],
 "Soil_Moisture": [30, 26, 35, 33, 29],
 "Yield_kg": [3200, 3000, 3500, 3400, 3100]
}
df = pd.DataFrame(data)
df
```

```python
from sklearn.linear_model import LinearRegression
X = df[["Rainfall_mm", "Soil_Moisture"]]
y = df["Yield_kg"]
model = LinearRegression()
model.fit(X, y)
```

```python
predicted_yield = model.predict(X)
predicted_yield
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
# Training data: past seasons
rainfall = np.array([[200], [190], [220], [250]])
yield_kg = np.array([3200, 3350, 3500, 3400])
# Train supervised model
model = LinearRegression()
model.fit(rainfall, yield_kg)
# Predict yield for new season
model.predict([[254]])
```

```python
from sklearn.cluster import KMeans
import pandas as pd
# Soil data without labels
soil_data = pd.DataFrame({
"pH": [6.5, 6.8, 7.2, 7.5, 6.6],
"Moisture": [30, 35, 45, 50, 32]
})
# Group soil types automatically
kmeans = KMeans(n_clusters=2, random_state=0)
soil_data["Group"] = kmeans.fit_predict(soil_data)
soil_data
```
